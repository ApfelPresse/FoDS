Consider	the	following	2-class	classification	problem:	Class	0 occurs	with	probability	p,	class	1	occurs	with	probability	1 − p.	For	i = 0,1 let	$x_i$ ∼ N($\mu, \sigma_i^2$) be	a	one-dimensional,	normally	distributed	random	variable with	class	label	i.	The	Bayes	classifier	with	zero-one	loss	assigns the	class	i to	an	$x \in R$ with	the	maximum	
posterior	probability	$P(y=i|x)$ .

a) Show	that	the	decision	boundary	of	the	Bayes	classifier	can	be	described	by	the	zero	of	a	polynomial	of	maximum	degree	2.

$ P(y=0|x) = P(y=0|x) $

$ ln(P(y=0|x)) - ln(P(y=1|x)) = 0 $

$ ln(P(x|y=0)P(y)) - ln(P(x|y=1)P(y)) = 0 $

$ ln(P(x|y=0)p) - ln(P(x|y=1)(1-p)) = 0 $

$ ln(P(x|y=0)p) - ln(P(x|y=1)(1-p)) = 0 $

$ -\frac{n}{2}ln(2\pi\sigma^2_0) + lnp - \frac{1}{2\sigma^2_0}\Sigma(x_i^2 - 2x_i\mu_0 + \mu_0^2) + lnp + \frac{n}{2}ln(2\pi\sigma^2_1) + ln(1-p) + \frac{1}{2\sigma^2_1}\Sigma(x_i^2 - 2x_i\mu_1 + \mu_1^2) + ln(1-p) = 0$

$ C - \frac{1}{2\sigma^2_0}\Sigma(x_i^2 - 2x_i\mu_0 + \mu_0^2) + \frac{1}{2\sigma^2_1}\Sigma(x_i^2 - 2x_i\mu_1 + \mu_1^2) = 0$

$ e = (1...1)^T $

$ C - \frac{1}{2\sigma^2_0}(e^TX^TX - 2e^TX\mu_0 + \mu_0^2) + \frac{1}{2\sigma^2_1}(e^TX^TX - 2e^TX\mu_1 + \mu_1^2) = 0$

b) In	which	case	is	the	degree	1?
Specify	the	decision	rule	for	this	case. What	is	the	influence	of	the	priors	here?

$ \sigma^2_0 = \sigma^2_1 = \sigma^2 $

$ C - (e^TX^TX - 2e^TX\mu_0 + \mu_0^2) + (e^TX^TX - 2e^TX\mu_1 + \mu_1^2) = 0$

$ C - 2e^TX\mu_0 + \mu_0^2 + 2e^TX\mu_1 + \mu_1^2 = 0$


c) In	which	case	is	the	degree	0?	Specify	the	decision	rule for	this	case. What	is	the	influence	of	the	priors	here?

$ \sigma^2_0 = \sigma^2_1 = \sigma^2 $

$ \mu_1 = \mu_0 = \mu $

# 5.4

In	this	exercise, we	illustrate	decision	boundaries	of	a	Bayes	classifier	with	normally	distributed	likelihoods.
For	 i = 0,1 let	 $x_i$ ~ $Ν(\mu_i ,\Sigma_i)$ 2-dimensional,	 normally	 distributed	 random	 variables	 with	 class	 label y = i and	prior	 P($y_i$) . Plot	data	points (numpy.random.multivariate_normal)	of	the	corresponding	classes	and	the	class	regions	in	different	colors	for	the	Bayes	classifier	with	zero-one	loss	for	the	following	parameters. Each plot	should	contain	900	data	points	in	total:

```python
size=900

def plot_bayes(mean_0,mean_1,cov_0,cov_1,p1,p2,size):
    dist_0 = np.random.multivariate_normal(mean=mean_0,cov=cov_0,size=(size//2))
    dist_1 = np.random.multivariate_normal(mean=mean_1,cov=cov_1,size=(size//2))

    x, y = np.mgrid[-7.5:12:.01, -8:8:.01]
    pos = np.empty(x.shape + (2,))
    pos[:, :, 0] = x; pos[:, :, 1] = y

    rv_0 = multivariate_normal(mean_0, cov_0)
    rv_1 = multivariate_normal(mean_1, cov_1)

    dens = rv_0.pdf(pos).copy()
    dens[p2*rv_0.pdf(pos) > p1*rv_1.pdf(pos)] = 0
    dens[p2*rv_0.pdf(pos) < p1*rv_1.pdf(pos)] = 1

    plt.figure(figsize=(12,8))
    plt.contourf(x, y, dens)
    plt.scatter(dist_0[:,0],dist_0[:,1], label="Class 0")
    plt.scatter(dist_1[:,0],dist_1[:,1], label="Class 1")
    plt.legend()

plot_bayes([0,0],[4,2],[[4,0],[0,1]],[[4,0],[0,1]],0.5,0.5,size)

plot_bayes([0,0],[4,2],[[4,0],[0,1]],[[4,0],[0,1]],0.3,0.6,size)

plot_bayes([0,0],[2,1],[[1,0],[0,1]],[[2,0],[0,2]],0.5,0.5,size)

plot_bayes([0,0],[2,1],[[1,0.5],[0.5,1]],[[1,0.5],[0.5,1]],0.25,0.75,size)

#plot_bayes([0,0],[2,1],[[1,2.1],[2,1]],[[1,0.5],[0.5,1]],0.5,0.5,size)

plot_bayes([0,0],[2,1],[[4,0.5],[0.5,1]],[[1,0.5],[0.5,4]],0.5,0.5,size)

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

def plot_bayes(mean_0,mean_1,mean_2,cov_0,cov_1,cov_2,p0,p1,p2,size):
    dist_0 = np.random.multivariate_normal(mean=mean_0,cov=cov_0,size=(size//3))
    dist_1 = np.random.multivariate_normal(mean=mean_1,cov=cov_1,size=(size//3))
    dist_2 = np.random.multivariate_normal(mean=mean_2,cov=cov_2,size=(size//3))
    
    min_x = np.minimum(dist_0[:,0].min(),dist_1[:,0].min())
    min_x = np.minimum(min_x,dist_2[:,0].min())
    
    max_x = np.maximum(dist_0[:,0].max(),dist_1[:,0].max())
    max_x = np.maximum(max_x,dist_2[:,0].max())
    
    min_y = np.minimum(dist_0[:,1].min(),dist_1[:,1].min())
    min_y = np.minimum(min_y,dist_2[:,0].min())
    
    max_y = np.maximum(dist_0[:,1].max(),dist_1[:,1].max())
    max_y = np.maximum(max_y,dist_2[:,0].max())

    x, y = np.mgrid[-1:max_x+3:.01, min_y-3:max_y+3:.01]
    pos = np.empty(x.shape + (2,))
    pos[:, :, 0] = x; pos[:, :, 1] = y

    rv_0 = multivariate_normal(mean_0, cov_0)
    rv_1 = multivariate_normal(mean_1, cov_1)
    rv_2 = multivariate_normal(mean_2, cov_2)

    dens = rv_0.pdf(pos).copy()
    dens[(p0*rv_0.pdf(pos) > p1*rv_1.pdf(pos)) & (p0*rv_0.pdf(pos) > p2*rv_2.pdf(pos))] = 0
    dens[(p1*rv_1.pdf(pos) > p0*rv_0.pdf(pos)) & (p1*rv_1.pdf(pos) > p2*rv_2.pdf(pos))] = 3
    dens[(p2*rv_2.pdf(pos) > p0*rv_0.pdf(pos)) & (p2*rv_2.pdf(pos) > p1*rv_1.pdf(pos))] = 6
    
    plt.figure(figsize=(12,8))
    plt.contourf(x, y, dens)
    plt.scatter(dist_0[:,0],dist_0[:,1], label="Class 0",marker='x')
    plt.scatter(dist_1[:,0],dist_1[:,1], label="Class 1",marker='^')
    plt.scatter(dist_2[:,0],dist_2[:,1], label="Class 2")
    plt.legend()

plot_bayes([0,0],[-2,4],[2,-2],[[2,0.5],[0.5,1]],[[1,0],[0,8]],[[1,0.25],[0.25,2]],0.33,0.33,0.33,900)
```
